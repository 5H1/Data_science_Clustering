```python
df=pd.read_csv("EastWEstAirlines.csv")
#Declaring X and Y variable
X = df.iloc[:,1:]
```

```python
X
```

```python
##THE DATA WHICH IS WITHOUT ID COLUMN IS SHOWN AS X
```

```python

from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)

SS_X.head()
```

```python
#APPLYING PCA AND IMPLEMENTING
```

```python
from sklearn.decomposition import PCA
pca = PCA()
```

```python
pc = pca.fit_transform(SS_X)
pc = pd.DataFrame(pc)
pc.head()

pc.iloc[:,0].var()
pc.iloc[:,1].var()
pc.iloc[:,2].var()

for i in range(0,10):
    print("variance: ",pc.iloc[:,i].var())

pd.DataFrame(pca.explained_variance_ratio_)

sum(pca.explained_variance_ratio_*100)
```

```python
X1=pc.iloc[:,0:4]
```

```python
X_12
```

```python
#implementing agglomerative clustering with only 2 columns from the pca data set
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
scores = []

for i in range(2,10):
    cluster = AgglomerativeClustering(n_clusters=i,linkage='single')
    X["Agg"] = cluster.fit_predict(X_12)
    Y = X["Agg"]
    scores.append(silhouette_score(X,Y))
    
    
import matplotlib.pyplot as plt
plt.scatter(range(2,10),scores)
plt.plot(range(2,10),scores,color='red')
plt.xlabel("cluster numbers")   
plt.ylabel("silhoutte scores")
plt.show()
```

```python
#==================================================================
# K - means clustering
#==================================================================

import pandas as pd
df = pd.read_csv("EastWestAirlines.csv")
df

X = df.iloc[:, 1:]
    
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=6,n_init=20)

df["kmeans"] = kmeans.fit_predict(SS_X)
df["kmeans"].value_counts()
```
